Import Libraries

```python
import pandas as pd
```

Load the Dataset

```python
# Adjust the path if your file is elsewhere
df = pd.read_csv("marketing_campaign.csv", sep="\t")
df.head()
```

Handle Missing Values

```python
# Fill missing values in 'Income' with median
df['Income'].fillna(df['Income'].median(), inplace=True)
```

Remove Duplicate Rows

```python
df.drop_duplicates(inplace=True)
```

Standardize Text Columns

```python
df['Education'] = df['Education'].str.strip().str.lower()
df['Marital_Status'] = df['Marital_Status'].str.strip().str.lower()
```

Convert Date Format

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], format='%d-%m-%Y')
```

Rename Columns

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

Confirm Data Cleaning

```python
df.info()
df.head()
```

Export Cleaned Data

```python
# Run this only once if needed
!pip install pandas openpyxl

# Save to CSV
df.to_csv("marketing_campaign_cleaned.csv", index=False)

# Optional: Save to Excel
df.to_excel("marketing_campaign_cleaned.xlsx", index=False)

print("✅ Data cleaned and exported.")
```
